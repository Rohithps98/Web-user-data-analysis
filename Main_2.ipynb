```python
import gc
import csv
import pandas as objpandas
import numpy as objnumpy
import matplotlib.pyplot as objplt
import seaborn as objsns
from sklearn.preprocessing import LabelEncoder
```

```python
#Load the data-set
dataset = objpandas.read_csv('/content/drive/MyDrive/weblog-analysis/Processed_12.csv', encoding='unicode_escape', low_memory=False) 

#Print the count of rows and coulmns in csv file
print("Dimensions of Dataset: {}".format(dataset.shape))
```

```python
# Dropped all the Null, Empty, NA values from csv file 
new_dataset = dataset.dropna(axis=0, how='any') 

print("Dimensions of Dataset after Pre-processing : {}".format(new_dataset.shape))
```

```python
collected = gc.collect()

print("Garbage collector: ", collected);
```

```python
#Data conversion for Url Column
filter_dataset_url = new_dataset.drop_duplicates(subset = ["Url"])

print("Dimensions of Dataset after Filtering : {}".format(filter_dataset_url.shape))
```

```python
Y = filter_dataset_url.iloc[:,3:4];

url_types = ();
url_types = list(url_types)

for i in range(len(Y)):
	url_types.append(Y.values[i]);

url_types = tuple(url_types)

print("url_types: ", url_types)
```

```python
filter_dataset_url_types = objpandas.DataFrame(url_types, columns=['Url_Types'])

# creating instance of labelencoder
labelencoder = LabelEncoder()

# Assigning numerical values and storing in another column
filter_dataset_url_types['Url_Category'] = labelencoder.fit_transform(filter_dataset_url_types['Url_Types'])

print("filter_dataset_url_types: ", filter_dataset_url_types);
```

```python
filter_dataset_url_types.to_csv('/content/drive/MyDrive/weblog-analysis/Processed_13.csv', encoding='utf-8')

#Data conversion for Agent IP Column
filter_dataset_ip = new_dataset.drop_duplicates(subset = ["Agent IP"])

print("Dimensions of Dataset after Filtering : {}".format(filter_dataset_ip.shape))
```

```python
Y = filter_dataset_ip.iloc[:,5:6];

ip_types = ();
ip_types = list(ip_types)

for i in range(len(Y)):
	ip_types.append(Y.values[i]);

ip_types = tuple(ip_types)

print("ip_types: ", ip_types)
```

```python
filter_dataset_ip_types = objpandas.DataFrame(ip_types, columns=['Agent_IP_Types'])

# creating instance of labelencoder
labelencoder = LabelEncoder()

# Assigning numerical values and storing in another column
filter_dataset_ip_types['Agent_IP_Category'] = labelencoder.fit_transform(filter_dataset_ip_types['Agent_IP_Types'])

print("filter_dataset_ip_types: ", filter_dataset_ip_types);
```

```python
filter_dataset_ip_types.to_csv('/content/drive/MyDrive/weblog-analysis/Processed_14.csv', encoding='utf-8')

# Using DataFrame.insert() to add a column
new_dataset.insert(8, "Url_Category", labelencoder.fit_transform(new_dataset['Url']), True)
new_dataset.insert(9, "Agent_IP_Category", labelencoder.fit_transform(new_dataset['Agent IP']), True)

#Remove dulicate values from url and agent ip
new_dataset_1 = new_dataset.drop_duplicates(subset = ["Url", "Agent IP"])

#Save the data to csv file
new_dataset_1.to_csv('/content/drive/MyDrive/weblog-analysis/Train.csv', encoding='utf-8')

print("Process Completed")
```
