```python
import csv
import re
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
Row_list=[]
dirc="/content/drive/MyDrive/input"
rowCount = 0
ProcessRowCount = 0
print("Process Started!")
Spider_list = ["googlebot","baidu","msn bot","bingbot","yandex","soso","exabot","sogou","google plus share",
               "facebook external hit","google feedfetcher","mi12bot","yahoo! slurp","ips-agent","panscient",
               "meanpathbot","scoutlet","exabot","archive.org_bot","twitterbot","surveybot","blexbot","qwant",
               "obot","linkdex","seznambot","applebot","alhitbot","spbot","webagent.wise-guys","plukkie","la_archive",
               "bot","spider","wget"]
SpiderRow_list =[]
Googlebot_list = []
Image_list =[]
SpiderCount = {}
ProcessedRowCount = 0
header_list = ['Date & Time', 'User IP', 'Method', 'Url', 'Query String', 'Agent IP', 'Browser Type', 'OS', 'Login Type', 'HTTP Status', 'Test1', 'Win 32 Status', 'Sub Status']
Row_list.append(header_list)

for file in os.listdir(dirc):
    print("Reading Input files from the directory!")
    filepath = dirc + "/" + file
    with open(filepath,newline='\n',encoding='latin1') as f:
        print("Processing file", file)
        #if rowCount == 5:
        #    break
        reader = csv.reader(f)        
        for row in reader:
            rowCount = rowCount + 1
            #if rowCount == 3:
            #    break
            if (not re.search("\.(?i:jpg|gif|png|css|js|jpeg|ajax)",str(row)) and not re.search("HEAD",str(row))):
                array = str(row).split(' ')   
                pagearr = array[4].split('/')
                PageName = pagearr[len(pagearr)-1]
                Status = int(array[len(array)-4])
                TimeSpent = int(array[len(array)-1][:-2])
                #if not Status.isnumeric():
                #    Status = array[12]
                if int(Status) < 200 or int(Status) > 299:
                    continue
                if not ".asp" in PageName:
                    continue
                    PageName = pagearr[len(pagearr)-1]
                date = array[0][2:]
                time = array [1]
                ip = array[2]
                method = array[3]
                Page = array[4]
                uriquery = array[5]
                UserAgentIP = array[8]
                UserAgent = array[9].replace(",","_")
                UserAgent1 = UserAgent + array[10].replace(",","_")
                substatus= int(array[len(array)-3])
                win32status = int(array[len(array)-2])
                datetime = date + " "+time
                browserName = UserAgent.split('/')[0]
                mobile = 0;
                if 'Android' in UserAgent1:
                    browserOS = 0 #Android
                    mobile = 1; #mobile login
                elif 'iPhone' in UserAgent1:
                    browserOS = 1 #iPhone
                    mobile = 1; #mobile login
                else:
                    browserOS = 2 #Web
                    mobile = 0; #web login
                for spi in Spider_list:
                    if spi in UserAgent1.lower():
                        if spi == "googlebot":
                            rowlist_2 = [datetime,ip,method,Page,uriquery,UserAgentIP,browserName,browserOS,mobile,Status,substatus,win32status,TimeSpent]
                            Googlebot_list.append(rowlist_2)
                        if spi in SpiderCount:
                            SpiderCount[spi] += 1
                        else:
                            SpiderCount[spi] = 1        
                        rowlist_1 = [datetime,ip,method,Page,uriquery,UserAgentIP,browserName,browserOS,mobile,Status,substatus,win32status,TimeSpent]
                        SpiderRow_list.append(rowlist_1)
                        continue
                if any(ele in UserAgent1 for ele in Spider_list):
                    continue
                #print("DateTime:",datetime)
                #print("IP:",ip)
                #print("method:",method)
                #print("Page:",Page)
                #print("uriquery:",uriquery)
                #print("UserAgentIP:",UserAgentIP)
                #print("UserAgent:",UserAgent)
                #print("UserAgent1:",UserAgent1)
                #print("Browser Name:",browserName)
                #print("Browser OS:",browserOS)
                #print("Mobile User:",mobile)
                #print("Status:",Status)
                #print("substatus:",substatus)
                #print("win32status:",win32status)
                #print("TimeSpent:",TimeSpent)
                ProcessRowCount = ProcessRowCount + 1
                rowlist = [datetime,ip,method,Page,uriquery,UserAgentIP,browserName,browserOS,mobile,Status,substatus,win32status,TimeSpent]
                Row_list.append(rowlist)
            else:
              # Splitting the ROW data into attributes
                array = str(row).split(' ')   
                pagearr=array[4].split('/')
                PageName=pagearr[len(pagearr)-1]
                Status = int(array[len(array)-4])
                TimeSpent = int(array[len(array)-1][:-2])
                PageName = pagearr[len(pagearr)-1]
                date = array[0][2:]
                time = array [1]
                ip = array[2]
                method = array[3]
                Page = array[4]
                uriquery = array[5]
                UserAgentIP = array[8]
                UserAgent = array[9].replace(",","_")
                UserAgent1 = UserAgent + array[10].replace(",","_")
                substatus= int(array[len(array)-3])
                win32status = int(array[len(array)-2])
                datetime = date + " "+time;
                browserName = UserAgent.split('/')[0];
                mobile = 0;
                if 'Android' in UserAgent1:
                    browserOS = 0 #Android
                    mobile = 1 #mobile login
                elif 'iPhone' in UserAgent1:
                    browserOS = 1 #iPhone
                    mobile = 1 #mobile login
                else:
                    browserOS = 0 #Web
                    mobile = 0; #web login
                rowlist = [datetime,ip,method,Page,uriquery,UserAgentIP,browserName,browserOS,mobile,Status,substatus,win32status,TimeSpent]
                Image_list.append(rowlist)
 
print("Process Ended!")
```
